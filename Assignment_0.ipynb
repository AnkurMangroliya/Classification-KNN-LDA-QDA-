```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mlt
```

```python
c = pd.read_csv('circles0.3.csv')
h = pd.read_csv('halfkernel.csv')
m = pd.read_csv('moons1.csv')
s = pd.read_csv('spiral1.csv')
t33 = pd.read_csv('twogaussians33.csv')
t42 = pd.read_csv('twogaussians42.csv')
```

```python
colors = {0: 'green', 1: 'blue'}

# Scatter plot implementation with circle dataset
plt.scatter(x=c['x1'], y=c['x2'], c=c['label'].map(colors))

plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
